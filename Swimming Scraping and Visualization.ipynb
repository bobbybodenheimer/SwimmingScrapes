```python
def process_file(file,column_names):
    d = pd.read_csv(file,header=None,names=column_names)
    d['Time']=d['Time'].apply(process_time)
    d['Date']=pd.to_datetime(d['Date'])
    return d
```

```python
def process_time(x):
    x = x.replace(' (r)','')
    x = x.replace(' (e)','')
    if ":" in x:
        time = x.split(":")
        return 60.0*float(time[0]) + float(time[1])
    else:
        return float(x)
```

```python
mb = process_file('100m-WR-back-M.csv',['Time','Swimmer','Date','Place'])
wb = process_file('100m-WR-back-W.csv',['Time','Swimmer','Date','Place'])
```

```python
mb = pd.read_csv('100m-WR-back-M.tsv',names=['Time','Swimmer','Date','Place'],sep='\t')
mb.head()
```

```python
def process_frame(d):
    d['Time']=d['Time'].apply(process_time)
    d['Date']=pd.to_datetime(d['Date'])
    return d
```

```python
mf['Event']=pd.Series('freestyle',index=mf.index)
```

```python
mf.head()
```

```python
mb['Event']=pd.Series('backstroke',index=mb.index)
mbr['Event']=pd.Series('breaststroke',index=mbr.index)
mfl['Event']=pd.Series('butterfly',index=mfl.index)
wf['Event']=pd.Series('freestyle',index=wf.index)
wb['Event']=pd.Series('backstroke',index=wb.index)
wbr['Event']=pd.Series('breaststroke',index=wbr.index)
wfl['Event']=pd.Series('butterfly',index=wfl.index)
mf=mf.rename(columns = {'Name':'Swimmer'})
wf=wf.rename(columns = {'Name':'Swimmer'})
```

```python
wr = pd.concat((mf[['Time','Date','Swimmer','Event']],mb[['Time','Date','Swimmer','Event']],mbr[['Time','Date','Swimmer','Event']],mfl[['Time','Date','Swimmer','Event']]),axis=0)
```

```python
wr.groupby('Event').plot(x='Date',y='Time')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(10,10))
wr.groupby('Event').plot(x='Date',y='Time',ax=axes)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(10,10))
for label,d in wr.groupby('Event'):
    d.plot(x='Date',y='Time',ax=axes,label=label)
plt.title("Men's Records")
plt.xlabel('Seconds')
plt.ylabel('Year')
```

```python
import numpy as np
wr['date_delta'] = (wr['Date'] - wr['Date'].min())  / np.timedelta64(1,'D')
```

```python
import statsmodels.formula.api as sm
for label, d in wr.groupby('Event'):
    results = sm.ols('Time ~ date_delta', d).fit()
    print('Event: ' + label)
    print(results.params)
    print(results.summary())
```
